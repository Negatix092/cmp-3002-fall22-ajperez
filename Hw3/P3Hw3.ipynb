## 3. Compare quick sort and merge sort

1. Generate 1000 random permutations of arrays of integers between 1 to 10000.

```python
def lists(amount, length):#function to generate 1000 arrays
    t = [];
    for i in range(amount):#for loop that generate 1000 arrays
        array = []
        for j in range(length):#for loop that stores 1000 arrays 
            #with random numbers between 1 and 10000
            array.append(np.random.randint(1,10001))
        t.append(array)
    
    r = tuple(t)#tuple to store the arrays
    return r
lists(5,6)
```

2. Execute merge sort and quick sort for each of the permutations

```python
def Quick_Sort(array):
    if len(array) <= 1:
        return array #easy case
    
    pivot = array.pop(0)#choosing first element as pivot
    left_array = []#creating the subarrays to store elements
    right_array = []#greater or smaller than the pivot
    
    for i in array:
        if i <= pivot:
            left_array.append(i)#element is smaller than the pivot
        else:
            right_array.append(i)#element is greater than the pivot
            
    #calling the function recursively and combining the result (concatenate)
    return Quick_Sort(left_array) + [pivot] + Quick_Sort(right_array)

def merge_sort(nums):
    if len(nums) <= 1:
        return nums
    pivot = int(len(nums) / 2)
    left = merge_sort(nums[0:pivot])
    right = merge_sort(nums[pivot:])
    return merge(left, right)


def merge(left, right):
    left_pointer = 0
    right_pointer = 0
    sorted_list = []
    while left_pointer < len(left) and right_pointer < len(right):
        if left[left_pointer] < right[right_pointer]:
            sorted_list.append(left[left_pointer])
            left_pointer += 1
        else:
            sorted_list.append(right[right_pointer])
            right_pointer += 1
    
    sorted_list.extend(left[left_pointer:])
    sorted_list.extend(right[right_pointer:])
    
    return sorted_list
```

```python
r = lists(1000,6)
t1 = time()

for x in r[:]:
    Quick_Sort(x)
    
total1 = (time() - t1)*10**6
print("total time:", str(total1))
```
